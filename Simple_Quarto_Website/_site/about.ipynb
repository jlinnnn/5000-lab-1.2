```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = 2 * x + 1  # y = 2x + 1

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='y = 2x + 1')  # 'b-' means blue solid line

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Basic Line Plot')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
