```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = 6 * x + 15
plt.figure(figsize=(10, 6))
plt.plot(x, y, '--r', label='y = 6x + 15') 

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Basic Line Plot')
plt.legend()

plt.show()
```
